```python
import numpy as mp
import pandas as pd 
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
def category(name):
    c = 0
    for i in df['Category']:
        if i == name:
            c += 1
    return c
```

```python
category('COMICS')
```

```python

def ratcount(nam):
    rat = list(df['Rating'])
    c=0
    d=0
    for j in df['Category']:
        
        if j == nam:
            if rat[c] > 4.5 :
                d+=1
        c+=1            
    print(nam, d)
    
```

```python
ratcount('COMICS')
```
